## Feature Engineering

#### Crearemos una variable que nos diga si el cliente va más de 35 meses en la empresa (probar número de meses)

```python
rango_tiempo_clientes = {range(1,13) : '0-1año', range(13,25) : '1-2años', #Creo un diccionario para reemplazar intervalos
     range(25,37) : '2-3años', range(37,49) : '3-4años',
     range(49, 79) : '5 a más'}

df['AñosCliente'] = df['MesesCliente'].apply(lambda x: next((v for k, v in rango_tiempo_clientes.items() if x in k), 0 ))
```

```python
rango_monto_total = {range(0,1801):'monto_bajo',range(1801,2301):'monto_intermedio',range(2301,4001):'monto_int_alto',
                    range(4001,9000):'monto_alto'}

df['Rango_monto_Cargado_total'] = df['MontoCargadoTotal'].apply(lambda x: 
                                                        next((v for k,v in rango_monto_total.items() if round(x) in k),0))
```

```python
rango_monto_mensual = {range(0,41):'bajo',range(41,61):'int',range(61,86):'int_alto',range(86, 240):'alto'}

df['Rango_monto_Cargado_mensual'] = df['MontoCargadoMes'].apply(lambda x:
                                                               next((v for k,v in rango_monto_mensual.items() if round(x) in k),0))
```

### Hacer clusters también es una forma de Feature Engineering

- Notemos que la mayoría de datos que tenemos son datos categóricos. Dado que el algoritmo KMeans está hecho en base a medias, no sería tan conveniente usarlo como método. Para estos casos una de las alternativas es usar KModes, donde el cálculo se usa en base a las modas y no a las medias. 

```python
from kmodes import kmodes
```

```python
# Elegiremos las variables categóricas que tenemos en nuestro modelo, exceptuando la variable objetivo
kmodes_variables = ['Sexo', 'AdultoMayor', 'Socio', 'Dependientes', 'MesesCliente',
       'ServicioTelefonico', 'LineasMultiples', 'ServicioInternet',
       'SeguridadOnline', 'RespaldoOnline', 'ProteccionDispositivo',
       'SoporteTecnico', 'TransmisionTV', 'TransmisionPeliculas',
       'TerminoContrato', 'FacturacionElectronica', 'MetodoPago','Rango_monto_Cargado_total',
                    'Rango_monto_Cargado_mensual']
#Llmamos a nuestro modelo y especificamos sus parámetros
clust_modes = kmodes.KModes(n_clusters=3,init='Huang', n_init=5)
#Entrenamos y etiquetamos nuestros registros según el cluster de modas
clusters_modes = clust_modes.fit_predict(df[kmodes_variables])
```

```python
# El cluster de modas irá como variable predictora dentro de nuestra data.
df['clusters_modes'] = clusters_modes
```

### Transformando Datos:

- Encodeando variables categóricas con LabelEncoder.

```python
from sklearn.preprocessing import LabelEncoder
cat_cols = [col for col in df.columns if df[col].dtype == 'O']

le = LabelEncoder()

for cat_col in cat_cols:
    le.fit(df[cat_col].values)
    df[cat_col] = le.transform(df[cat_col])
```

### Midiendo el performance de nuestros modelos:

El accuracy score no es la única forma de medir nuestros modelos.

```python
from sklearn.metrics import roc_curve

fig = plt.figure(figsize=(8,6))

for i in tabla_resultados.index:
    plt.plot(tabla_resultados.loc[i]['fpr'], 
             tabla_resultados.loc[i]['tpr'], 
             label="{}, AUC={:.3f}".format(i, tabla_resultados.loc[i]['auc']))
    
plt.plot([0,1], [0,1], color='orange', linestyle='--')

plt.xticks(np.arange(0.0, 1.1, step=0.1))
plt.xlabel("Ratio de Falsos Positivos", fontsize=15)

plt.yticks(np.arange(0.0, 1.1, step=0.1))
plt.ylabel("Ratio de Verdaderos Positivos", fontsize=15)

plt.title('Análisis de la curva ROC', fontweight='bold', fontsize=15)
plt.legend(prop={'size':12}, loc='lower right')

plt.show()
```
